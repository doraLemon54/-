```python
import requests
import lxml.html
import pandas as pd
import time
```

```python
for code in range(10001, 190000):
    url = 'https://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword={}&target=after={}'
    res = requests.get(url.format(code,1))
    element = lxml.html.fromstring(res.text)
    
    try: 
        total = int(element.find('.//strong[@class="c_88 fs_11"]').text_content())
    except:
        continue
        
    last_pages = int(total / 10)
    
    if last_pages == 0:
        continue
        
    reviews = []
    start = time.time()

    for page in range(1, last_pages):
            res = requests.get(url.format(code, page))
            element = lxml.html.fromstring(res.text)
            for e in element.xpath('.//table[@class="list_netizen"]//tbody//tr'):
                star = e.find('.//td[@class="point"]').text_content()
                comment = e.find('.//td[@class="title"]').text_content()
                comment = comment.strip().replace("\t", "").replace("\r","").split("\n")[4]
                title = e.find('.//td[@class="title"]').find('a').text_content()
                id_wrtdate = e.find('.//td[@class="num"]').text_content()
                id = id_wrtdate[:-8]
                wrtdate = id_wrtdate[-8:]
                star = int(star)
                reviews.append([id,wrtdate,title,star,comment])
            if page % 100 == 0:
                print('=>', end='')
                time.sleep(3)
    review = pd.DataFrame(reviews, columns=['id','wrtdate','title','star','comment'])
    review.to_csv('data/{}_reviews.csv'.format(code), index=False)
    print('{} : review collecting {} page time : {}'.format(code, last_pages, round(time.time()-start)/60.0, 2))
```
